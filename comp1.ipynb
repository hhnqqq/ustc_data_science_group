```python
import pandas as pd
train_df = pd.read_json('D:\\workspace\\data_analysis\\data\\comp1\\train.json')
train_df
```

```python
train_df['entities'] = train_df.user.apply(lambda x: x['entities'])
```

```python
count_dict={}
for idx,row in train_df.iterrows():
    for key, value in row['user'].items():
        if key not in count_dict:
            count_dict[key] = 1
        else:
            count_dict[key] += 1
count_dict
```

```python
def get_full_df(df):
    for idx, row in df.iterrows():
        for key, value in row['user'].items():
            df.at[idx, key] = value
    df.drop(columns=['user'], inplace=True)
    return df
train_df = get_full_df(train_df)
train_df
```

```python
train_df.describe()
```

```python
bot_len = 0
human_len = 0
for idx,row in train_df.iterrows():
    if row['label'] == 'bot':
        bot_len += len(row['screen_name'])
    else:
        human_len += len(row['screen_name'])
print('the average name len of bots are {} and average len of human are {}'.format(bot_len/train_df[train_df['label']=='bot'].shape[0],human_len/train_df[train_df['label']=='human'].shape[0]))
```

```python
train_df['name_length'] = train_df.screen_name.apply(lambda x: len(x))
train_df.drop(columns=['screen_name'],inplace=True)
train_df
```

```python
train_df['created_year'] = train_df.created_at.apply(lambda x: int(str(x).split('-')[0]))
train_df.followers_count = train_df.followers_count.apply(lambda x: int(x))
```

```python
import matplotlib.pyplot as plt


bot_accounts = train_df[train_df["label"] == 'bot']
non_bot_accounts = train_df[train_df["label"] == 'human']

plt.scatter(bot_accounts["created_year"], bot_accounts.followers_count, color='blue', label='Bot Accounts')
plt.scatter(non_bot_accounts["created_year"], non_bot_accounts.followers_count, color='red', label='Non-Bot Accounts')

plt.legend()
plt.title("Bot Accounts vs. Non-Bot Accounts by Creation Year")
plt.xlabel("Creation Year")
plt.ylabel("Account Index")

plt.show()
```

```python
lang_dict = {lang:id for id, lang in enumerate(train_df.lang.unique())}
train_df.lang = train_df.lang.apply(lambda x: lang_dict[x])
train_df.lang
```

```python
bot_counts = train_df[train_df['label'] == 'bot'].groupby('lang').size()
total_counts = train_df.groupby('lang').size()
lang_bot_ratio_df = pd.DataFrame({'Bot_Count': bot_counts, 'Total_Count': total_counts})
lang_bot_ratio_df['Bot_Ratio'] = lang_bot_ratio_df['Bot_Count'] / lang_bot_ratio_df['Total_Count']
lang_bot_ratio_df
```

```python
train_df['have_url'] = train_df.url.apply(lambda x: not pd.isna(x))
train_df.have_url
```
